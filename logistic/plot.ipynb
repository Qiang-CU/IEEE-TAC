```python
import os 
import numpy as np
from util import plot_figure
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
color_list = sns.color_palette()
```

```python
data_dir = 'res/'
ss = 1

fig, ax = plt.subplots(figsize=(5.5, 5))

plot_csgd = plot_figure(algo_name=f'csgd', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_csgd.plot_lines(ax, color_list[0], label='CSGD', legend=True, plot_star=True)

plot_dsgd_homo = plot_figure(algo_name=f'dsgd-homo-R', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_dsgd_homo.plot_lines(ax, color_list[1], label='Homo-DSGD', legend=True, plot_star=True)

plot_gt_homo = plot_figure(algo_name=f'gt-homo-R', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_gt_homo.plot_lines(ax, color_list[3], label='Homo-GT', legend=True, plot_star=True)

plot_edas_homo = plot_figure(algo_name=f'edas-homo-R', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_edas_homo.plot_lines(ax, color_list[6], label='Homo-EDAS', legend=True, plot_star=True)

plot_ex_homo = plot_figure(algo_name=f'ex-homo-R', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_ex_homo.plot_lines(ax, color_list[8], label='Homo-ExDi', legend=True, plot_star=True)

plot_dsgd_er_homo = plot_figure(algo_name=f'dsgd-homo-ER', sub_sample=ss, dir = data_dir , log_flag=True, metric='mse')
plot_dsgd_er_homo.plot_lines(ax, color_list[9], label='Homo-DSGD-ER', legend=True, plot_star=True)

plt.grid(True)
ax.set_xlim([1e0, 1e5])
ax.set_ylabel(r"$\|\bar{\theta}^{t}-\theta^{\star}\|^2$", fontsize = 18)
ax.set_title('Homo. Data', fontsize = 18)
# ax.set_xscale('linear')
# ax.ticklabel_format(style='sci', axis='x', scilimits=(0,1))
```
