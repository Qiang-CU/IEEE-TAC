```python
from util import plot_figure
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
color_list = sns.color_palette()
line_style = ['--', ':', '-']
```

```python
path = 'res/'

fig, ax = plt.subplots(figsize=(5.5,4.3))
log_flag = True
plot_instance = plot_figure(algo_name='dectd',  dir=path, log_flag_=log_flag, sub_sample=1)
plot_instance.plot_lines(ax, color_list[0], label=f'Decentralized TD(0)', plot_star=True)

plot_instance = plot_figure(algo_name='centd',  dir=path, log_flag_=log_flag, sub_sample=1)
plot_instance.plot_lines(ax, color_list[1], label=f'Centralized TD(0)', plot_star=True)

# plt.xlim([1e0, 1e5])
# plt.ylim([1e-4, 1e3])
plt.ylabel(r"$\|\bar{\theta}_t-\theta^\star\|^2$", fontsize = 20)
ax.set_xlabel(r"Iterations $t$", fontsize=18)
plt.ticklabel_format(axis='x', style='sci', scilimits=(4, 4))
plt.xscale('log')
plt.yscale('log')
plt.legend(loc=3, prop={'size': 14})
plt.grid()
plt.savefig('./fig/' + 's2-itrate.pdf', bbox_inches='tight')
plt.show()
```
